```python
# A json module is built-in package in python. 
# Since our dataset is in json format, json module is imported from python to perform json operations like load data, read, write.
# A nltk module helps us to preprocess the training data.
# A word_tokenize transforms sentence into words/tokens.
import json
import nltk
from nltk.tokenize import word_tokenize
```

```python
# Here greetings dataset is loaded.
with open('research_dataset.json', 'r') as f:
    intents = json.load(f)
```

```python
# Here we are looping through the greetings json file
# And adding the tokenized words from inquiry and response to the list.
tokenized_words = []
tokenized_inquiry_words = []
tokenized_response_words = []
for intent in intents['intents']:
    for inquiry,response in zip(intent['inquiry'],intent['responses']):
        tokenized_inquiry_words.extend(word_tokenize(inquiry))
        tokenized_response_words.extend(word_tokenize(response))
tokenized_words.extend(tokenized_inquiry_words)
tokenized_words.extend(tokenized_response_words)
print(tokenized_words)
```

```python
# To perform Frequency on the words (i.e., word count), we need to import "FreqDist" module from "nltk" package.
from nltk.probability import FreqDist
```

```python
# Showing the overall word count
# Passing the tokenized words list to the "FreqDistribution" module to check for each word's count.
# Then printing top 30 words with the respective frequency.
print("Word count:",len(tokenized_words))
freq_dist = FreqDist(tokenized_words)
freq_dist.most_common(30)
```

```python
# In order to plot and show the plots, we import "matplotlib.pyplot" module.
import matplotlib.pyplot as plt
freq_dist.plot(30)
```

```python
# Removing punctuations by using "isalpha()" method which checks whether the input string contains alphabets or not.
words_with_no_punctuation = []

for i in tokenized_words:
    if i.isalpha():
        words_with_no_punctuation.append(i.lower())
print(words_with_no_punctuation)
```

```python
# Passing the tokenized words list to the "FreqDistribution" module to check for each word's count.
# Then printing top 30 words with the respective frequency.

freq_Dist = FreqDist(words_with_no_punctuation)
final_freq_list = freq_Dist.most_common(30)
final_freq_list
```

```python
#plotting the frequency of words after removing the punctuation.
freq_Dist.plot(30)
```

```python
# A wordcloud represents the frequency of words in a colorful picture. The bigger the size of the words, the higher it's frequency.
# To perfrom wordcloud, we import "WordCloud" module from "wordcloud" package.
# To show the plots in python, we need to import "matplotlib.pyplot" module.
from wordcloud import WordCloud

import matplotlib.pyplot as plt

# Since the output frequency distribution list has tuples in it, we need to convert the tuples into a dictionaries.
# The converted dictionaries is given to "generate_from_frequencies" function in "WordCloud" module.
# Later the figure is plotted with customized size.

for i,j in final_freq_list:
    freq_dictionary = dict((i,j) for i,j in final_freq_list)
    
for i,j in freq_dictionary.items():
    freq_dictionary[i] = j

wordcloud = WordCloud().generate_from_frequencies(freq_dictionary)
plt.figure(figsize=(12,12))
plt.imshow(wordcloud)

plt.axis("off")
plt.show()
```
